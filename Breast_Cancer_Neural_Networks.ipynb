```python
# Importing the Dependencies
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# data collection and processing
```

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer_dataset   # prints all the features and target
```

```python
# Creating variable for data frame
# loading the data to a data frame
breast_cancer_data = pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
breast_cancer_data.head()
```

```python
# adding the target colm to data frame
breast_cancer_data['label'] = breast_cancer_dataset.target
```

```python
# printing last 5 rows of the data frame
breast_cancer_data.tail()
```

```python
# num of rows and colms
breast_cancer_data.shape
```

```python
# getting some info about the data
breast_cancer_data.info()
```

```python
# No Missing Values
breast_cancer_data.isnull().sum()
```

```python
# Statistical Measures about the data
breast_cancer_data.describe()
```

```python
# Checking the distribution of Target variable
breast_cancer_data['label'].value_counts()
```

```python
# 1 ---> Benign
# 0 ---> Malignant 
```

```python
# Grouping based on label
# for  0 ---> Malignant  valueas are higher
breast_cancer_data.groupby('label').mean()
```

```python
# separating the features and targets
```

```python
X = breast_cancer_data.drop(columns='label',axis=1)
Y = breast_cancer_data['label']
```

```python
X
```

```python
Y
```

```python
# Splitting the data into training and testing data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Standardize the data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
# fitting and transforming the data >>> only training data 
X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
X_train_std
```

```python
 #busilding the Neural Network
```

```python
!pip install tensorflow
```

```python
# importing the tensor flow and keras
import tensorflow as tf
# to get same accuracy score after every run
tf.random.set_seed(3)  
from tensorflow import keras
```

```python
# Setting up the layers of NN

model = keras.Sequential([
                         keras.layers.Flatten(input_shape=(30,)),
                         keras.layers.Dense(20,activation='relu'),
                         keras.layers.Dense(2,activation='sigmoid')
])


# 1.# Flatten() convert the data into 1-d array 
# # i.e no of nuerons in i/p layer = no of features in data set

# 2. 2nd layer is hidden layer
# 3. num of neurons in o/p layers = num of classes you have
# Creating 2 neurons in o/p layer
```

```python
# Compiling the NN

model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])

# sparse_categorical_crossentropy used  for label encoding 
```

```python
# Training the NN

history = model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)

# validation_split is same as train_test_split
# epochs is how many times the model has to go through the data
```

```python
# Visualizing accuracy and loss
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
# Accuracy of the model on test data
```

```python
loss,accuracy = model.evaluate(X_test_std,Y_test)

# model uses X_test_std and predicts the labels .those are compared with Y_test
```
